```python
import pyreadstat
import pandas as pd
from scipy import stats
# 读取.sav文件
df, meta = pyreadstat.read_sav('国家认同数据清理后.sav')
# 假设我们关注的是第三年级的学生（假设年级用数字“2”表示）
specific_grade = 3
# 筛选出特定年级的数据
df_specific_grade = df[df['年级'] == specific_grade]
# 创建政治面貌和会以中国人自豪吗的交互表
cross_tab_specific_grade = pd.crosstab(df_specific_grade['政治面貌'], df_specific_grade['会以中国人自豪吗'])
# 进行卡方检验
chi2_grade, p_grade, dof_grade, expected_grade = stats.chi2_contingency(cross_tab_specific_grade)

print(chi2_grade, p_grade)
print(cross_tab_specific_grade)
```

这个卡方检验的结果显示，特定年级学生的政治面貌和是否以中国人自豪之间的关系并不具有统计学意义。具体来说，卡方值为16.61，p值为0.165。根据通常的显著性水平（通常为0.05），p值大于0.05意味着我们无法拒绝原假设，即观察到的结果可能是由随机因素引起的，而非真正的关联。因此，根据这个p值，我们无法得出特定年级学生的政治面貌和是否以中国人自豪之间存在显著关联的结论。
这意味着在这个特定年级的学生中，政治面貌和是否以中国人自豪之间的关系可能是随机的，而非由于学生的政治面貌导致的。这个结果可能需要更多的数据或者其他分析方法来进一步确认是否存在真正的关联。
